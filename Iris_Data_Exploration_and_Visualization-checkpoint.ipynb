```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set better visualization style
sns.set(style="whitegrid")

print("Libraries imported successfully!")
```

```python
# Load the Iris dataset
df = pd.read_csv("Iris.csv")  # Make sure Iris.csv is in the same folder as this notebook

# Show first 5 rows
df.head(5)
```

```python
# Shape of dataset (rows, columns)
print("Dataset shape:", df.shape)

# Column names
print("\nColumns in dataset:", df.columns)
```

```python
# Info about dataset (data types, non-null values)
df.info()

# Summary statistics
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop ID column if exists
if "Id" in df.columns:
    df = df.drop("Id", axis=1)

df.head(5)
```

```python
# Scatter plot Sepal Length vs Petal Length
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="SepalLengthCm", y="PetalLengthCm", hue="Species", palette="Set1")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
# Histograms for all numerical features
df.hist(figsize=(10,8))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop("Species", axis=1))
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
# Pairplot to see relationships between features
sns.pairplot(df, hue="Species", palette="Set2")
plt.show()
```

```python
# Heatmap of correlations
plt.figure(figsize=(8,6))
sns.heatmap(df.drop("Species", axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
